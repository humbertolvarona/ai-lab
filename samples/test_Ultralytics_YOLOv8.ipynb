```python
from ultralytics import YOLO
```

```python
import matplotlib.pyplot as plt
import cv2
import glob
```

```python
model = YOLO("data/yolov8n.pt") 
```

```python
results = model("https://ultralytics.com/images/bus.jpg")
```

```python
results[0].save(filename="data/output.jpg")
```

```python
print("Image saved in: data/output.jpg")
```

```python
img = cv2.imread("data/output.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Resultado de YOLOv8")
plt.show()
```
